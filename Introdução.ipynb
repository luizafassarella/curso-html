**Algoritmo Textual:** Codificação

```
# ENTRADA DOS DADOS: Coeficientes A, B e C
A = int(input('Coeficiente A: '))
B = float(input('Coeficiente B: '))
C = float(input('Coeficiente C: '))


if A == 0 or (B == 0 and C == 0):
  print('Não é uma Equação do 2° Grau.')
else:
  delta = B ** 2 - 4 * A * C
  print(f'Valor de Delta: {delta: .2f}')
  if delta < 0:
    print('Não tem Raízes no conjunto dos números reais.')
  elif delta == 0:
    X1 = -B / (2 * A)
    print(f'Raiz X1 = X2 = {X1: .2f}')
  else:
    raizDelta = delta ** 0.5
    print(f'Valor da Raíz de Delta: {raizDelta: .2f}')
    X1 = (-B + raizDelta) / (2 * A)
    X2 = (-B - raizDelta) / (2 * A)
    print(f'Raiz X1 = {X1: .2f}')
    print(f'Raiz X2 = {X2: .2f}')
```

## **Exemplos de Algoritmos**:

### **Exemplo 1**:

Faça um algoritmo em Pythom que leia de um aluno (1) da academia:

$\longmapsto$  Medida da Cintura (em centímetros): MC

$\longmapsto$  Medida da Quadril (em centímetros): MQ

E exiba na tela seu $RCQ$  (Relação Cintura-Quadril). Sendo que:

$ R_{CQ}  =  \frac{MC}{MQ}* 100$

```
# ALGORITMO = ENTRADA (INPUT) + PROCESSAMENTO + SAÍDA (PRINT)
# ENTRADA (input): variáveis lidas do usuário
MC = float(input('Medida da Cintura (em centímetros): '))
MQ = float(input('Medida do Quadril (em centímetros): '))

#PROCESSAMENTO:
RCQ = MC / MQ * 100  # Em porcentagem: %

# SAÍDA (print): Valores (objetivo) exibido para usuário
print(f'𝑅𝐶𝑄 (Relação Cintura-Quadril): {RCQ: .1f}%')
```

### **Exemplo 2**:

Escrever um algoritmo em Python que leia de um (1) aluno sua nota do 1º e 2º bimestre e

```
# ALGORITMO: ENTRADA + PROCESSAMENTO + SAÍDA (OBJETIVO)
try:
  # ENTRADA:
  B1 = float(input('Nota [0-10] - 1º Bimestre: '))
  B2 = float(input('Nota [0-10] - 2º Bimestre: '))

  # PROCESSAMENTO
  MS = (B1 + B2) / 2

  # SAÍDA (OBJETIVO):
  print(f'Média Semestral: {MS: .1f}')
except:
  print('Erro nos dados de entrada.')
```

### **Exemplo 3**:

Escrever um algoritmo em Python que leia de uma (1) mercadoria:

*   Preço: R$
*   Quantidade: Unidade
*   Reajuste (Desconto): %

exbir na tela:


*   Total a pagar: sem desconto
*   Total a pagar: com desconto



```
# ALGORITMO = ENTRADA (INPUT) + PROCESSAMENTO + SAÍDA (PRINT)
try:
  preco = float(input('Preço (R$): '))
  quantidade = int(input('Quantidade (Unid.): '))
  reajuste = float(input('Reajuste (%): '))

  # Como calcular a porcentagem?
  total = preco * quantidade
  print(f'Total a pagar (sem desconto): R$ {total: .2f}')
  total = total - total * reajuste / 100
  # total = total * (1 - reajuste / 100)
  print(f'Total a pagar (com desconto: {reajuste: .1f} %): R$ {total: .2f}')
except Exception as ERRO:
  print(f'ERRO: {ERRO}')
```
